```python
def obj_to_float(val):
  if isinstance(val,str):
    if "cm" in val:
      val= val.replace("cm",'')
      return float(val)
    elif "\'" in val:
      val=val.split("\'")
      h=float(val[0])*30.48+float(val[1].replace('"',''))*(2.54)
      return h
    elif "kg" in val:
      val=val.replace("kg","")
      return float(val)
    elif "lbs" in val:
      val=float(val.replace("lbs",""))*(0.45359237)
      return val
    elif "K" in val:
      val = val.replace("K", "*1000")
      return eval(val)
    else:
      return float(val)
  else:
    return float(val)
```

```python
fifa_updated['Height'] =fifa_updated['Height'].apply(obj_to_float)
fifa_updated['Weight'] =fifa_updated['Weight'].apply(obj_to_float)
fifa_updated['Hits'] =fifa_updated['Hits'].apply(obj_to_float)
```

```python
fifa_updated.rename(columns={'↓OVA':'OVA'},inplace=True)
```

```python
def change_currency(val):
  if isinstance(val,str):
    if "K" in val:
      val= float(val.replace("K","").replace("€",""))*1000*90.57
      return val
    elif "M" in val:
      val=float(val.replace("M","").replace("€",""))*10e6
      return val
    elif "€" in val:
      val=float(val.replace("€",""))
      return val
    else:
      return float(val)
  else:
    return(val)
```

```python
fifa_updated['Value']=fifa_updated['Value'].apply(change_currency)
fifa_updated['Release Clause']=fifa_updated['Release Clause'].apply(change_currency)
fifa_updated['Wage']=fifa_updated['Wage'].apply(change_currency)
```
